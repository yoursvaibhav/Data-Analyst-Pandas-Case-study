## handling missing values

```python
#mode()[0] returns the most frequent value in each column.

mode_workclass = data['workclass'].mode()[0]  
mode_occupation = data['occupation'].mode()[0]
mode_country = data['native-country'].mode()[0]
```

```python
data['workclass_mode'] = data['workclass'].fillna(mode_workclass)
data['occupation_mode'] = data['occupation'].fillna(mode_occupation)
data['native-country_mode'] = data['native-country'].fillna(mode_country)

'''
df['workclass'].fillna(mode_workclass, inplace=True)
df['occupation'].fillna(mode_occupation, inplace=True)
df['native-country'].fillna(mode_native_country, inplace=True)

'''
```
